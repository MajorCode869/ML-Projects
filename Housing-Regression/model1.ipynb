```python



plt.figure(figsize=(16,6))
heatmap=sns.heatmap(cr,vmin=-1,vmax=1,annot=True,cmap='BrBG')
heatmap.set_title('Pearson Correlation Heatmap',fontdict={'fontsize':18},pad=12)

```

```python
# remove independent correlated features
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.9,random_state=42)
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
lreg=LinearRegression()
lreg.fit(X_train,y_train)
y_pred=lreg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
print(lreg.coef_)
```

```python
print(lreg.intercept_)
```

```python
lreg.get_params()
```

## Plotting Results 

```python
# checking the residuals if they are normally distributed
residuals=y_test-y_pred
sns.displot(residuals,kind="kde")
```
